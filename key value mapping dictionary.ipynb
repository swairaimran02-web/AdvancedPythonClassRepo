```python
# --- 1. Merging Databases ---
inventory_tech = {"Laptop": 1200, "Mouse": 25}
inventory_office = {"Desk": 300, "Chair": 150, "Mouse": 30}

# The | operator merges dictionaries.
# Note: If keys overlap (like 'Mouse'), the second dictionary wins.
full_inventory = inventory_tech | inventory_office

print("--- Merged Inventory ---")
print(full_inventory)


# --- 2. Safe Retrieval with .get() ---
# Using inventory["Tablet"] would crash the script if it doesn't exist.
# .get() allows for a "fallback" value.
item_search = "Tablet"
price = full_inventory.get(item_search, "Not in Database")

print(f"\nSearch Result for '{item_search}': {price}")


# --- 3. Dictionary Comprehension (Flash Sale) ---
# Goal: Create a new dictionary with prices halved.
# Syntax: {key: new_value for (key, value) in dictionary.items()}

flash_sale_prices = {item: price / 2 for (item, price) in full_inventory.items()}

print("\n--- 50% OFF FLASH SALE ---")
for item, sale_price in flash_sale_prices.items():
    print(f"{item}: ${sale_price:.2f}")
```

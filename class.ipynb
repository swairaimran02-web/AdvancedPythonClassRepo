```python

class UserProfile:
    def __init__(self, username, age, gender):
        self.username = username
        self.__age = age
        self.gender = gender

    def display_info(self):
        print("Username:", self.username)
        print("Age:", self.age)
        print("Gender:", self.gender)

    def greet(self):
        print(f"Hello {self.username}, welcome!")

    def activity(self):
        print(f"{self.username} is using the system.")
class cUserProfile(UserProfile):
    def __init__(self, username, age, gender, subscription_type, points):

        super().__init__(username, age, gender)
        self.subscription_type = subscription_type
        self.points = points

    def show_subscription(self):
        print(f"Subscription Type: {self.subscription_type}")

    def show_points(self):
        print(f"Reward Points: {self.points}")

    def activity(self):
        print(f"{self.username} is a premium user enjoying exclusive features!")
```

```python
user1 = UserProfile("Ali", 20, "Male")
print("---- Normal User ----")
user1.display_info()
user1.greet()
user1.activity()

print("\n---- Child User ----")
user2 = cUserProfile("Sara", 22, "Female", "Gold", 1500)
user2.display_info()
user2.greet()
user2.activity()  # Polymorphism (overridden method)
user2.show_subscription()
user2.show_points()
```

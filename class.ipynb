```python
class email:
    def __init__(self,Email):
        self.Email = Email
    def account(self,Name,Password):
        print(Name," ",Password)
class emailAccount(email):
    def username(self,Name):
        print(Name)
    def password(self,Password):
        print(Password)
eml=emailAccount()
eml.username("John")
eml.password("sik")
eml.account("John","sik")
```

```python
class pclass:
    def parentClassFunction(self):
        print("Parent Class Function")
class cClass(pclass):
    def childClassFunction(self):
        print("Child Class Function")
child=cClass()
child.parentClassFunction()
```

```python
import random as rd
computer = rd.randint(0,100)
attempt=0
print("welcome to number game")
print("I am thinking of a number between 0 and 100")
while True:
    number=int(input("Enter a number: "))
    attempt+=1
    if number>computer:
        print(f"too high!")
    elif number<computer:
        print("too low!")
    else:
        print(f"Correct!")
        print(number)
        break

print(f"you guessed the no in {attempt } attempts")
```

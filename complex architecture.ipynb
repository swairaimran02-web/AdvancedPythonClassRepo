```python
S# --- 1. The Nested Architecture ---
# Dictionary (Warehouses) -> List (Inventory) -> Dictionary (Parts)
warehouses = {
    "North_Hub": [
        {"id": 101, "name": "sensor", "qty": 50},
        {"id": 102, "name": "actuator", "qty": 20}
    ],
    "East_Hub": [
        {"id": 101, "name": "sensor", "qty": 35},
        {"id": 103, "name": "motor", "qty": 10}
    ],
    "South_Hub": [
        {"id": 101, "name": "sensor", "qty": 100},
        {"id": 102, "name": "actuator", "qty": 15}
    ]
}

# --- 2. The Logic: Cross-Warehouse Totals ---
target_part = "sensor"
total_quantity = 0

# Outer loop: Iterates through each warehouse name (the keys)
for location in warehouses:
    # Inner loop: Iterates through the list of part dictionaries
    for part in warehouses[location]:
        if part["name"] == target_part:
            total_quantity += part["qty"]

# --- 3. Report Output ---
print(f"--- Global Inventory Report ---")
print(f"Targeting Part: {target_part.upper()}")
print(f"Total Stock Found: {total_quantity} units")
print(f"Status: {'Restock Required' if total_quantity < 200 else 'Stock Levels Normal'}")
```

```python
import random
li = ["rock","paper","sci"]
computerChoice = random.choice(li)
userChoice = input("Choose rock, paper or sci")
print("you choose:",userChoice)
print("computer choose:",computerChoice)
if userChoice.lower() == "rock" and computerChoice == "paper" or userChoice.lower() == "paper" and computerChoice == "sci" or userChoice.lower() == "sci" and computerChoice == "rock":
    print("----You Loose----")
elif userChoice.lower()=="rock" and computerChoice=="rock" or userChoice.lower()=="paper" and computerChoice=="paper" or userChoice.lower()=="sci" and computerChoice=="sci":
    print("----Draw----")
else:
    print("----you won----")
```

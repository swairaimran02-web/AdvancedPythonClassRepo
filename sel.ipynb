```python
import time
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By


browser = uc.Chrome(version_main=146)
browser.get("https://orteil.dashnet.org/cookieclicker/")

time.sleep(10)



browser.find_element(By.ID, "langSelect-EN").click()
time.sleep(5)




cookie = browser.find_element(By.ID, "bigCookie")


cookiecount = browser.find_element(By.ID, "cookies")


cursorbutton = browser.find_element(By.ID, "product0")

multipliers = {"K":1_000,"M":1_000_000,"B":1_000_000_000,"T":1_000_000_000_000}


for i in range(500):
    cookie.click()
    time.sleep(0.05)


    cookiecounttext = cookiecount.text.split(" ")[0].replace(",", "")
    if cookiecounttext[-1] in multipliers:
        count = int(float(cookiecounttext[:-1]) * multipliers[cookiecounttext[-1]])
    else:
        count = int(float(cookiecounttext))


    pricetext = cursorbutton.find_element(By.CLASS_NAME, "price").text.replace(",", "")
    if len(pricetext) > 0 and pricetext[-1] in multipliers:
        price = int(float(pricetext[:-1]) * multipliers[pricetext[-1]])
    elif pricetext.isdigit():
        price = int(pricetext)
    else:
        price = 0
    if count >= price and price > 0:
        cursorbutton.click()

```

```python
# Unit Conversion System

while True:
    print("\n--- Unit Conversion System ---")
    print("1. Length")
    print("2. Temperature")
    print("3. Weight")
    print("4. Time")
    print("5. Exit")

    choice = input("Select a category (1-5): ")

    # LENGTH CONVERSION
    if choice == "1":
        value = float(input("Enter value: "))
        print("Units: meter, kilometer, inch, foot")
        unit_from = input("Enter current unit: ").lower()
        unit_to = input("Convert into: ").lower()

        # Convert everything to meters first
        if unit_from == "meter":
            meters = value
        elif unit_from == "kilometer":
            meters = value * 1000
        elif unit_from == "inch":
            meters = value * 0.0254
        elif unit_from == "foot":
            meters = value * 0.3048
        else:
            print("Invalid unit!")
            continue

        # Convert meters to target unit
        if unit_to == "meter":
            result = meters
        elif unit_to == "kilometer":
            result = meters / 1000
        elif unit_to == "inch":
            result = meters / 0.0254
        elif unit_to == "foot":
            result = meters / 0.3048
        else:
            print("Invalid unit!")
            continue

        print("Result:", result, unit_to)

    # TEMPERATURE CONVERSION
    elif choice == "2":
        value = float(input("Enter value: "))
        print("Units: celsius, fahrenheit")
        unit_from = input("Enter current unit: ").lower()
        unit_to = input("Convert into: ").lower()

        if unit_from == "celsius" and unit_to == "fahrenheit":
            result = (value * 9/5) + 32
        elif unit_from == "fahrenheit" and unit_to == "celsius":
            result = (value - 32) * 5/9
        elif unit_from == unit_to:
            result = value
        else:
            print("Invalid conversion!")
            continue

        print("Result:", result, unit_to)

    # WEIGHT CONVERSION
    elif choice == "3":
        value = float(input("Enter value: "))
        print("Units: kilogram, gram, pound")
        unit_from = input("Enter current unit: ").lower()
        unit_to = input("Convert into: ").lower()

        # Convert to kilograms
        if unit_from == "kilogram":
            kg = value
        elif unit_from == "gram":
            kg = value / 1000
        elif unit_from == "pound":
            kg = value * 0.453592
        else:
            print("Invalid unit!")
            continue

        # Convert kg to target
        if unit_to == "kilogram":
            result = kg
        elif unit_to == "gram":
            result = kg * 1000
        elif unit_to == "pound":
            result = kg / 0.453592
        else:
            print("Invalid unit!")
            continue

        print("Result:", result, unit_to)

    # TIME CONVERSION
    elif choice == "4":
        value = float(input("Enter value: "))
        print("Units: second, minute, hour")
        unit_from = input("Enter current unit: ").lower()
        unit_to = input("Convert into: ").lower()

        # Convert to seconds
        if unit_from == "second":
            sec = value
        elif unit_from == "minute":
            sec = value * 60
        elif unit_from == "hour":
            sec = value * 3600
        else:
            print("Invalid unit!")
            continue

        # Convert seconds to target
        if unit_to == "second":
            result = sec
        elif unit_to == "minute":
            result = sec / 60
        elif unit_to == "hour":
            result = sec / 3600
        else:
            print("Invalid unit!")
            continue

        print("Result:", result, unit_to)

    # EXIT
    elif choice == "5":
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Please try again.")
```

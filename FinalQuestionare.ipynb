```python
import pandas as pd
import random
import os

# read question file
questionare = pd.read_csv("QuestionareDummy.csv")

# load previous scores
if os.path.exists("student_score.csv"):
    userDataFrame = pd.read_csv("student_score.csv")
else:
    userDataFrame = pd.DataFrame({"Names":[], "Score":[]})

def start_quiz(name):

    score = 0

    random_questions = random.sample(range(len(questionare)), 10)

    for q_no in random_questions:

        question = questionare.loc[q_no]

        print("\nQuestion:", question["Question"])
        print({f"A.question['QA']}")
        print({f"B., question['QB']}")
        print({f"C., question["C"]}")
        print({f"D., question["D"]}")

        ans = input("Enter answer (A/B/C/D): ").upper()

        if ans == question["Answer"].upper():
            print("Correct!")
            score += 1
        else:
            print("Wrong Answer! Game Over")
            break

    print("Your Score:", score)

    global userDataFrame

    new_row = pd.DataFrame({"Names":[name], "Score":[score]})

    userDataFrame = pd.concat([userDataFrame,new_row], ignore_index=True)

    userDataFrame.to_csv("student_score.csv", index=False)


while True:

    print("\n1.Take Quiz")
    print("2.Exit")

    ch = input("Enter choice: ")

    if ch == "1":

        name = input("Enter your name: ")

        if name in list(userDataFrame["Names"]):
            print("You already attempted the quiz")
        else:
            start_quiz(name)

    elif ch == "2":
        break

    else:
        print("Invalid Choice")
```

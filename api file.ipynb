```python
import requests

url = 'https://example.com'
headers = {
    'Content-Type': 'application/json'
}
#//thing we send and it give us response
payload = {'title':'I am in love with someone.'}

# Using json= automatically encodes the dict and sets Content-Type header
response = r.post(url, json=payload, headers=headers)

print(response.status_code)
print(response.json())
```

```python
import requests as r
login_url = 'https://api.escuelajs.co/api/v1/auth/login'

headers = {
    'Content-Type': 'application/json'
}

credentials = {
    "email": "john@mail.com",
    "password": "changeme"
}

response = r.post(login_url, json=credentials, headers=headers)

print("Login Status:", response.status_code)

login_data = response.json()
print("Login Response:", login_data)

access_token = login_data.get("access_token")
profile_url = "https://api.escuelajs.co/api/v1/auth/profile"

new_headers = {
    "Authorization": f"Bearer {access_token}"
}

profile_response = r.get(profile_url, headers=new_headers)

print("Profile Status:", profile_response.status_code)
print("Profile Data:", profile_response.json())
profile_data = profile_response.json()
```

```python
import os
image_url = profile_data.get("avatar")

folder_name = "images"
os.makedirs(folder_name, exist_ok=True)

image_path = os.path.join(folder_name, "profile1.jpg")

img_data = r.get(image_url).content

with open(image_path, "wb") as f:
    f.write(img_data)

print("Image saved at:", image_path)


profile_data["image_path"] = "https://imgur.com/LDOO4Qs.jpg"

df = pd.DataFrame([profile_data])

print(df)
```
